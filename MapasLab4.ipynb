```python
import folium
from folium.plugins import MiniMap
from branca.element import Figure

popuptext = '<b>Guanacaste</b>'
my_map = folium.Map(location=[10.2856822,  -84.9008912], zoom_start=16, tiles='Stamen Terrain')
folium.Marker(location=[10.2856822, -84.9008912], popup=popuptext, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.2856822, -84.90008912], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Guanacaste").add_to(my_map)

minimap = MiniMap(tile_layer='Stamen Terrain', position='bottomright')
my_map.add_child(minimap)
my_map.save('C:folium.html')
my_map
```

```python
import folium
from folium.plugins import MiniMap

# Crear un mapa
my_map = folium.Map(location=[10.2856822, -84.9008912], zoom_start=16, tiles='Stamen Terrain')

# Punto 1 - Liberia
popuptext1 = '<b>Liberia</b>'
folium.Marker(location=[10.2856822, -84.9008912], popup=popuptext1, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.2856822, -84.90008912], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Liberia").add_to(my_map)

# Punto 2 - Benjamin Nuñez (Lagunilla)
popuptext2 = '<b>Campus Benjamin Nuñez</b>'
folium.Marker(location=[9.969871325165677, -84.12899216164543], popup=popuptext2, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[9.969871325165677, -84.12899216164543], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Campus Benjamin Nuñez").add_to(my_map)

# Punto 3 - Omar Dengo (Heredia)
popuptext3 = '<b>Campus Omar Dengo</b>'
folium.Marker(location=[9.998950159864458, -84.11141972817722], popup=popuptext3, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[9.998950159864458, -84.11141972817722], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Campus Omar Dengo").add_to(my_map)

# Punto 4 - Nicoya
popuptext4 = '<b>Nicoya</b>'
folium.Marker(location=[10.134536373145897, -85.44676975598405], popup=popuptext4, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.134536373145897, -85.44676975598405], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Nicoya").add_to(my_map)

# Punto 5 - Campus Coto
popuptext5 = '<b>Campus Coto</b>'
folium.Marker(location=[8.573397780822622, -82.87225737559501], popup=popuptext5, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[8.573397780822622, -82.87225737559501], color="purple", weight=6, fill=True, fill_color="red", fill_opacity=0.5, tooltip="Campus Coto").add_to(my_map)

# Agregar el minimapa
minimap = MiniMap(tile_layer='Stamen Terrain', position='bottomright')
my_map.add_child(minimap)

# Guardar el mapa en un archivo HTML
my_map.save('folium.html')

# Mostrar el mapa
my_map
```
